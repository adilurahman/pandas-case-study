#8) Handle missing values in 'Revenue (Millions)' and 'Metascore


````python
df.info()
````

````python
df['Revenue (Millions)'].fillna(value=df['Revenue (Millions)'].mean(), inplace=True)
#missing values filled with mean
df['Metascore'].fillna(value=df['Metascore'].median(), inplace=True)
print(df)
#Missing values filled with median
df.info()
````

#14) Group by Director: avg Rating, Revenue

#16) Create a new column Profitability:


````python
df["Profitability"]=df["Revenue (Millions)"] /df["Runtime (Minutes)"]


df
````

#17) Create a column that categorizes movies by rating:


````python
df["Category"]=df["Rating"].apply(lambda x:"poor" if x <6  else None)
df.loc[(df["Rating"]>=6) & (df["Rating"]<7.5 ),"Category"]="avarage"
df.loc[df["Rating"]>7.5,"Category"]="excelent"
df[["Title","Rating","Category"]]
````

#18) Create a custom function to score movies based on:


````python
def score_movie(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['Score'] = df.apply(score_movie, axis=1)
df[['Title', 'Score']].head(3)
````
