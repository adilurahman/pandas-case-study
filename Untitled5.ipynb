```python
import numpy as np
import pandas as pd
import sqlite3
import requests
```

#Step 1: Load SpaceX Launch Data from API


```python
url1=' https://api.spacexdata.com/v4/launches'
responce= requests.get(url1)
data=responce.json()
#step 2 :convert to data frime
lounches_df=pd.DataFrame(data)
lounches_df
```

```python
lounches_df=lounches_df[['name','date_utc','success','details','rocket']]
lounches_df
```

```python
lounches_df['date_utc']=pd.to_datetime(lounches_df['date_utc'])
lounches_df['year']=lounches_df['date_utc'].dt.year
lounches_df
```

#Step 2: Load Rocket Metadata


```python
url2='https://api.spacexdata.com/v4/rockets'
request= requests.get(url2)
data=request.json()
rockets_df=pd.DataFrame(data)
rockets_df
```

```python
rockets_df=rockets_df[['id','name','type','active','stages']]
rockets_df
```

#Step 3: Merge Launch and Rocket Data


```python
merged_df = pd.merge(
    lounches_df,
    rockets_df,
    left_on='rocket',
    right_on='id',
    how='left'
)
merged_df
```

#Step 4: Add Simulated Country Information

```python
countries=['USA', 'Russia', 'India', 'China', 'France']
merged_df['country']=np.random.choice(countries,size=len(merged_df))
merged_df
```

#Step 5: Store Merged Data in SQLite3

```python
conn=sqlite3.connect('launche.db')
```

```python
merged_df.to_sql('launche',conn,index=False)
```

```python
df_from_db=pd.read_sql('SELECT *FROM launche LIMIT 5',conn)
df_from_db
```

#Step 6: Run SQL Queries on the Data to analyze

```python

cursor=conn.cursor()
cursor.execute("""
select country,count(*) as lounches_count
from launche
group by country
order by lounches_count desc""")
```

```python
rows=(cursor.fetchall())
for row in rows:
  print(row)
```

```python
cursor.execute("""
select year,count(*) as lounches_count
from launche
group by year
order by lounches_count desc limit 1""")
rows=(cursor.fetchall())
for row in rows:
  print(row)
```

```python
cursor.execute("""
select name_x,count(*) as lounches_count
from launche
group by name_x
order by lounches_count desc limit 5""")
rows=(cursor.fetchall())
for row in rows:
  print(row)
```
